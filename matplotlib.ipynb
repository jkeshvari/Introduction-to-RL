```python
### Matplotlib
import matplotlib.pyplot as plt
import numpy as np
xpnt = np.array([0,6])
ypnt = np.array([0,250])
## plot as line
plt.plot(xpnt,ypnt)
plt.show()
## plot point to point
plt.plot(xpnt,ypnt,'o')
plt.show()
```

```python
## plot with more details
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'x', ls = '-.', lw = '1', color = 'r', ms = 5, mec = 'b', mfc = 'r' )
plt.show()

## ls
# ls = line style both can be written 
# 'solid' (default)	'-'	
# 'dotted'	':'	
# 'dashed'	'--'	
# 'dashdot'	'-.'	
# 'None'	'' or ' '

### What you can put in marker
# 'o'	Circle	
# '*'	Star	
# '.'	Point	
# ','	Pixel	
# 'x'	X	
# 'X'	X (filled)	
# '+'	Plus	
# 'P'	Plus (filled)	
# 's'	Square	
# 'D'	Diamond	
# 'd'	Diamond (thin)	
# 'p'	Pentagon	
# 'H'	Hexagon	
# 'h'	Hexagon	
# 'v'	Triangle Down	
# '^'	Triangle Up	
# '<'	Triangle Left	
# '>'	Triangle Right	
# '1'	Tri Down	
# '2'	Tri Up	
# '3'	Tri Left	
# '4'	Tri Right	
# '|'	Vline	
# '_'	Hline
```

```python
### Draw two graphs in one plot
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
x2 = np.array([0,1,2,4])
y2 = np.array([6,2,7,11])
plt.plot(x1,y1, label = 1)
plt.plot(x2,y2, label = 2)
##title & label

font1 = {'family' : 'serif' , 'color' : 'blue', 'size' : 20}
font2 = {'family' : 'serif' , 'color' : 'darkred', 'size' : 15}

plt.title("Sports Watch Data", fontdict = font1 , loc = 'left')

plt.xlabel("Average Pulse", fontdict = font2)

plt.ylabel("Calorie burnage", fontdict = font2)

plt.grid()

plt.legend()


plt.show()
```

```python
## Scatter plot
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,3,4,10,14,11,6,16,3,1])
y = np.array([15,13,20,45,60,34,5,73,55,31,27,14,12])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x,y,c = colors, s = 30)
# s = size
# c = color
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=100)
# print(x)
y = np.random.randint(100, size=100)

colors = np.random.randint(100, size=100)

size = 10*np.random.randint(100, size=100)
plt.scatter(x,y, c = colors, s = size, alpha = 0.5 , cmap='nipy_spectral')

plt.colorbar()

plt.plot()
```

```python
## Bar plot
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y)
plt.show()

plt.barh(x,y,height=0.5)
plt.show()
```

```python
## pie plot
mylabels = ["Apple","Banana","cherry","Dates"]
y = np.array([35,25,25,15])
exp = [0.1,0,0.1,0]
plt.pie(y,labels=mylabels,startangle=0,explode=exp)
plt.legend(loc = 'right', title = 'Fruits')

plt.show()
```
